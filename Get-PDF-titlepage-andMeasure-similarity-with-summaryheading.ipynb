```python
# Open a PDF document.
fp = open('/home/muhammadsiraj/Downloads/- Team KIET Stars - Alejandro Cremades - The Art of Startup Fundraising - Pitching Investors, Negotiating the Deal, and Everything Else Entrepreneurs Need to Know (2016).pdf', 'rb')
parser = PDFParser(fp)
document = PDFDocument(parser)
# Get the outlines of the document.
outlines = document.get_outlines()
myls = []
for (level,title,dest,a,se) in outlines:
    myls.append(title)
    print (level, title)
    
```
