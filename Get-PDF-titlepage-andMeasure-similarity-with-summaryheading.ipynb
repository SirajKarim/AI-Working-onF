```python
# Open a PDF document.
fp = open('/home/muhammadsiraj/Downloads/Natural Language Processing with Python.pdf', 'rb')
parser = PDFParser(fp)
document = PDFDocument(parser)
# Get the outlines of the document.
outlines = document.get_outlines()
myls = []
for (level,title,dest,a,se) in outlines:
    myls.append(title)
    print (level, title)
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import euclidean_distances
import numpy as np

myls.insert(0,'NLTK Roadmap')
print(myls[0])
mySentences = myls
vectorizer = CountVectorizer()
features =vectorizer.fit_transform(mySentences).todense()
print(vectorizer.vocabulary_)

for f in features:
    print(euclidean_distances(features[0],f))
```
